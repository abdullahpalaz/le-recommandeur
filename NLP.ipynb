```python
import pandas as pd
df = pd.DataFrame({'Songs' : songs,
                                'Lyrics' : lyrics}, 
                                columns=['Songs','Lyrics'])
```

```python
nltk.download('punkt')
df['Lyrics'] = df['Lyrics'].str.replace('\d+', '') #deleting numbers
df['Lyrics'] = df['Lyrics'].str.lower() #make all the letters lowercase
df['Lyrics'] = df.apply(lambda row: nltk.sent_tokenize(row['Lyrics']), axis=1) #tokenize the sentences
```

```python
df
```

```python
def punctuation_remove(token): #remove function for punctuations
    clean_words=[]
    for i in token:
        for j in s.punctuation:
            i=i.replace(j,'')
        clean_words.append(i)
    return clean_words
```

```python
df['Lyrics'] = df['Lyrics'].apply(punctuation_remove)
```

```python
def space_remove(token): #remove function for extra spaces
    non_space_list=[]
    for i in token:
        i = i.strip()
        non_space_list.append(i)
    return non_space_list
```

```python
df['Lyrics'] = df['Lyrics'].apply(space_remove)
```

```python
df
```

```python
df['Lyrics'] = df['Lyrics'].str.join(" ")
```

```python
mood = []
print("Negative  -   Positive")
for i in range(0,len(songs)):
  lyrical = df['Lyrics'][i]
  print(sid.polarity_scores(lyrical)['neg'] , "      ", sid.polarity_scores(lyrical)['pos'] )
  if(sid.polarity_scores(lyrical)['pos'] - sid.polarity_scores(lyrical)['neg'] > 0.05 ):
    truth = "Positive"
    mood.append(truth)
  else:
    truth = "Negative"
    mood.append(truth)
```

```python
df['Mood'] = mood
df = df.drop('Lyrics', axis=1)
df
```
